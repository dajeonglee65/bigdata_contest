```python
def tem_inout(ptnum, df, x, y):
    temin = ptnum + '_train.tem_in_' + y
    temout = ptnum + '_train.tem_out_loc1'
    var_tem_inout = 'tem_inout_' + y
    df[var_tem_inout] = df[temin]-df[temout]
    a = df.groupby(x)[var_tem_inout].agg([('평균','mean')]).reset_index()
    a[x] = a[x].astype('str')
    plt.bar(a[x], a['평균'])
```

```python
tem_inout('plant1', pt1, 'plant1_train.cond_loc1', 'loc1')
```

```python
tem_inout('plant1', pt1, 'plant1_train.cond_loc2', 'loc2')
```

```python
def coil_in_loc(ptnum, df, x, y):
    temin = ptnum + '_train.tem_in_' + y
    temcoil = ptnum + '_train.tem_coil_' + y
    newcoilin = 'coil_in_' + y
    df[newcoilin] = df[temin] - df[temcoil]
    coilinratio = 'coil_in_ratio_' + y
    df[coilinratio] = df[newcoilin]/df[temin]
    a = df.groupby(x)[coilinratio].agg([('평균','mean')]).reset_index()
    a[x] = a[x].astype('str')
    return plt.bar(a[x], a['평균'])
```
